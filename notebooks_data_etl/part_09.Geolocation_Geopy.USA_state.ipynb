```python
df_state_codes = pd.read_json("data/state_codes.json")
df_state_codes[:3]
```

```python
def unpack_locations(x,a=-2):
    "Check if is numeric"
    return ",".join([i for i in x['display_name'].split(",") if i.strip().isnumeric()==False][a:]).strip()

def get_country_from_dict(df):
    df["Country_Geopy"] = df.Country_location.apply(unpack_locations)
    df = df.set_index("Country_Geopy")
    return df

def check_and_compare(dfContryCode,seriesIndex):
    return dfContryCode.loc[seriesIndex]["code"]

def get_code(country_code,x):
    return  check_and_compare(country_code, x)
```

```python
state_code = get_country_from_dict(df_state_codes)
state_code[:3]
```

```python
state_code.rename(index={'District of Columbia, United States':'Washington, United States'},inplace=True)
state_code.loc['District of Columbia, United States'] = ["DC", 'District of Columbia', float("nan")]  
```

```python
df["State_Code"] = df.Country_Geopy.apply(lambda x: x != "United States" and x != "Wake Island, United States" and get_code(state_code,x))
```
